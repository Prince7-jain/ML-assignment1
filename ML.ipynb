Q> To plot the frequency distribution of the highest base of random strings.


```python
import random
import matplotlib.pyplot as plt
mapping_dict = dict(zip(range(0,10),range(0,10)))
mapping_dict
for k in mapping_dict.keys():

  mapping_dict[k] = str(mapping_dict[k])
mapping_dict
for i in range(0,26):
  mapping_dict[10+i] = chr(65+i)
mapping_dict
sample_size = 100000

random_numbers = list()

for i in range(0,sample_size):

  blank_str = str()

  random_digit = random.randint(0,35)
  blank_str = blank_str + mapping_dict[random_digit]

  random_digit = random.randint(0,35)
  blank_str = blank_str + mapping_dict[random_digit]

  for i in range(0,8):
    coin_toss = random.randint(0,1)

    if coin_toss == 1:

      random_digit = random.randint(0,35)
      blank_str = blank_str + mapping_dict[random_digit]

    else:
      break

  random_numbers.append(blank_str)
reverse_mapping_dict = dict()

for k in mapping_dict.keys():
  reverse_mapping_dict[mapping_dict[k]] = k
base_frequency = dict()

for blank_str in random_numbers:
  highest_digit = max(blank_str)
  highest_base = reverse_mapping_dict[highest_digit] + 1

  if highest_base in base_frequency.keys():
    base_frequency[highest_base] += 1/sample_size
  else:
    base_frequency[highest_base] = 1/sample_size
plt.bar(x=base_frequency.keys(),height=base_frequency.values())
```

**DOCUMANTATION OF ABOVE CODE =>**

This Program or code generates a large number of random strings and then analyzes the distribution of the highest base in each string.

**Modules:**
- random: Provides various functions for generating random numbers.
- matplotlib.pyplot: Provides a MATLAB-like interface for making plots and charts.

**Variables:**
- mapping_dict: A dictionary where the keys are numbers from 0 to 35 and the values are strings of digits (0-9) and uppercase letters (A-Z).
- sample_size: The number of random strings to generate.
- random_numbers: A list of randomly generated strings.
- reverse_mapping_dict: A dictionary where the keys are the values from mapping_dict and the values are the keys from mapping_dict.
- base_frequency: A dictionary where the keys are bases and the values are their frequencies in the randomly generated strings.

**Procedure:**
1. Initialize mapping_dict with keys and values as numbers from 0 to 9.
2. Convert the values in mapping_dict to strings.
3. Add entries to mapping_dict where the keys are numbers from 10 to 35 and the values are uppercase letters from A to Z.
4. Set sample_size to 100000.
5. Initialize an empty list random_numbers.
6. Generate sample_size number of random strings and append them to random_numbers. Each string starts with two characters followed by up to eight more characters based on a coin toss.
7. Initialize an empty dictionary reverse_mapping_dict.
8. Create a reverse mapping dictionary from mapping_dict.
9. Initialize an empty dictionary base_frequency.
10. Calculate the frequency of the highest base in each string in random_numbers and store the frequencies in base_frequency.
11. Create a bar plot of the frequencies in base_frequency.
